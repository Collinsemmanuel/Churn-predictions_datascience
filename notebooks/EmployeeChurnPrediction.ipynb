```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder, StandardScaler
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
#create a DataFrame from the given data
data = {
    'Name': ['Robin Hood', 'Arsene Wenger', 'Friar Tuck', 'Little John', 'Sam Allardyce', 'Dimi Berbatov'],
    'Department': [None, 'Bar', 'Foo', 'Foo', None, 'Foo'],
    'Manager': [None, 'Friar Tuck', 'Robin Hood', 'Robin Hood', None, 'Little John'],
    'Salary': [200, 50, 100, 100, 250, 50]
}
```

```python
df = pd.DataFrame(data)
```

```python
#encoding categorical variables
label_encoder = LabelEncoder()
df['Department'] = label_encoder.fit_transform(df['Department'].astype(str))
df['Manager'] = label_encoder.fit_transform(df['Manager'].astype(str))
```

```python
#Scaling the categorical features
scaler = StandardScaler()
df['Salary'] = scaler.fit_transform(df['Salary'].values.reshape(-1, 1))
```

```python
#Build a neural network model
model = Sequential([
    Dense(64, activation='relu', input_shape=(3,)),  # 3 input features after encoding
    Dense(32, activation='relu'),
    Dense(1, activation='sigmoid')
])
```

```python
# Compile the model
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
```
